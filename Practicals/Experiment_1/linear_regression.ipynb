```python
import numpy as np 
import pandas as pd
from sklearn.metrics import r2_score,accuracy_score
from sklearn.datasets import fetch_california_housing
```

```python
from sklearn.datasets import make_regression
import pandas as pd

# Create synthetic regression data
X, y = make_regression(
    n_samples=200,      # number of data points
    n_features=1,       # number of features
    noise=10,           # add some noise
    random_state=42
)

# Combine into a DataFrame (optional, for easy viewing)
df = pd.DataFrame(X, columns=['Feature'])
df['Target'] = y

print(df.head())
```

```python

class Lr():
  """OLR Odinary least Square Regression"""
  def __init__(self):
    self.m=None
    self.b=None

  def fit(self,x_train:np.array,y_train:np.array):

    N=sum((x_train-x_train.mean())*(y_train-y_train.mean()))
    D=sum((x_train-x_train.mean())*(x_train-x_train.mean()))

    m=N/D
    b=y_train.mean()-(m*x_train.mean())

    self.m=m
    self.b=b

  def predict(self,x_test:np.array):
    return self.m*x_test+self.b

  def accuracy(self,y_test:np.array,y_pred:np.array):
    return np.sum(y_test-y_pred)/len(y_test)
```

```python
#train test split 
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test=train_test_split(df['Feature'],df['Target'],test_size=0.2,random_state=42)
```

```python
lr=Lr()

lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)

lr.accuracy(y_test,y_pred)
```

## 2 or More Features

```python
from sklearn.datasets import make_regression
import pandas as pd

# Create synthetic regression data
X, y = make_regression(
    n_samples=500,      # number of data points
    n_features=2,       # number of features
    noise=10,           # add some noise
    random_state=42
)

# Combine into a DataFrame (optional, for easy viewing)
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Target'] = y

print(df.head())
```

```python

class Multipe_Lr():
  """MLR Odinary least Square Regression"""
  def __init__(self):
    self.coef=None
    self.intercept=None

  def fit(self,X_train,y_train):
    X_train=np.insert(X_train,0,1,axis=1)
    beta=np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
    self.intercept=beta[0]
    self.coef=beta[1:]
    return self.coef,self.intercept

  def predict(self,X_test):
    y_pred=np.dot(X_test,self.coef)+self.intercept
    return y_pred
  def accuracy(self,y_test,y_pred):
    return np.sqrt(np.sum((y_test-y_pred)**2)/len(y_test))
```

```python
#train test split 
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
mlr=Multipe_Lr()
mlr.fit(X_train,y_train)
y_pred=mlr.predict(X_test)
print(f"RMSE of Our Multiple Linear Regression :{mlr.accuracy(y_test,mlr.predict(X_test))}")
```
