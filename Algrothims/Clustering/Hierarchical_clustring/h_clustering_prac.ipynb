```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
file_name=r"C:\Users\Administrator\OneDrive\Desktop\Machine_learning_cs303\Algrothims\Clustering\hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv"
customer_data = pd.read_csv(file_name)
customer_data.head()
```

```python
data=customer_data.iloc[:,3:5]
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')

labels_=cluster.fit_predict(data)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=cluster.labels_, cmap='rainbow')
plt.show()
```
